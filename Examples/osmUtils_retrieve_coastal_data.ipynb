```python

import LMIPy as lmi #Docs https://lmipy.readthedocs.io/en/latest/quickstart.html#From-Political-Boundaries

params={
    'iso': 'ESP',
    'adm1': 6,
    'adm2': 2
}

sp_geom = lmi.Geometry(parameters=params)
sp_geom.map()
```

```python
# get the geometry
geometry = sp_geom.shape()[0]
geometry
```

```python
#retrieve data without metadata
osm_Data_coastline= osmu.OsmDownload(
    geometry,
    osm_type='coastline',
    includes_metadata=True
)
```

```python
coastline_gdf = osm_Data_coastline.get_osm_gdf()
coastline_gdf.head()
```

```python
#visualize the data retrieved
osmu.OsmVisualize(coastline_gdf, color='#1c16c7')
```

```python
clean_gdf = coastline_gdf[['geometry', 'id', 'timestamp']]
clean_gdf
```

```python
clean_gdf.dtypes
```

```python
clean_gdf.to_file(
    './data/coastal_data.json',
    driver='GeoJSON')
```
